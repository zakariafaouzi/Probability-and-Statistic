### Exo 1 (Simulation d’une Loi de Bernoulli)

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_bernoulli(p, n=10000):
    return np.random.binomial(1, p, n)

p=0.7
bernoulli_samples = simulate_bernoulli(p)

#La moyenne empirique
mean_empirical = np.mean(bernoulli_samples)
print(f"Moyenne empirique : {mean_empirical}")
print(f"La moyenne théorique égale: {p}")
```

### Exo 2 (Simulation d’une Loi Binomiale)

```python
def simulate_binomiale(n, p, essai = 1000):
    return np.random.binomial(n, p, essai)

p = 0.5
n = 10
binomial_samples = simulate_binomiale(n, p)

# Comparer les moyennes
mean_empirical = np.mean(binomial_samples)
print(f"Moyenne empirique : {mean_empirical}")
print(f"La moyenne théorique égale: {n*p}")

# Affichage de l'histogramme
plt.hist(binomial_samples, bins=np.arange(0, n+2)-0.5, density=True, label='Empirique')
plt.title('Distribution des échantillons binomiaux')
plt.xlabel('Nombre de succès')
plt.ylabel('Densité')
plt.xticks(range(n+1))
plt.grid()
plt.legend()
plt.show()
```

__Distribution :__ Si tu vois que les valeurs autour de 5 (c'est-à-dire 4, 5 et 6 succès) ont les barres les plus hautes, cela indique que ces résultats sont les plus probables, ce qui est attendu avec une probabilité de succès de 0.5 sur 10 essais.

__Écart :__ Si la barre pour 0 succès est très basse et celle pour 10 succès est également basse, cela renforce l'idée que les extrêmes sont moins probables avec cette configuration.

### Exercice 3 : Simulation d’une Loi G ́eom ́etrique

```python
def simulate_geometric(p, trials=5000):
    return np.random.geometric(p, trials)

p = 0.3
geometric_samples = simulate_geometric(p)

mean_empirical = np.mean(geometric_samples)
mean_theoretical = 1 / p

print(f"Moyenne empirique : {mean_empirical}")
print(f"Moyenne théorique : {mean_theoretical}")

# Affichage de l'histogramme
plt.hist(geometric_samples, bins=range(1, max(geometric_samples)+1), color='g')
plt.title('Distribution des échantillons géométriques')
plt.xlabel('Nombre d\'essais avant le premier succès')
plt.ylabel('Densité')
plt.grid()
plt.show()
```
