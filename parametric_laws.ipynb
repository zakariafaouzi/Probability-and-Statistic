```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.preprocessing import PowerTransformer
```

```python
# G√©n√©rer des donn√©es exponentielles
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000)

plt.figure(figsize=(12, 6))

# Histogramme des donn√©es initiales
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution avant transformation")
plt.xlabel("Valeurs")
plt.ylabel("Fr√©quence")

```

```python
# Appliquer la transformation Box-Cox (PowerTransformer avec m√©thode 'box-cox')
pt = PowerTransformer(method='box-cox')
data_boxcox = pt.fit_transform(data.reshape(-1, 1)).flatten()
```

```python
plt.subplot(1, 2, 2)
plt.hist(data_boxcox, bins=30, color='lightcoral', edgecolor='black')
plt.title("Distribution apr√®s Box-Cox")
plt.xlabel("Valeurs")
plt.ylabel("Fr√©quence")

plt.tight_layout()
plt.show()
```

```python
# V√©rification de la normalit√© avec le test de Shapiro-Wilk
stat_before, p_before = stats.shapiro(data)
stat_after, p_after = stats.shapiro(data_boxcox)

print(f"Test de Shapiro-Wilk avant transformation : p-value = {p_before:.4f}")
print(f"Test de Shapiro-Wilk apr√®s transformation : p-value = {p_after:.4f}")
```
